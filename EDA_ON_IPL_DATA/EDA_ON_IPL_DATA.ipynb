```python
# I want all the records where team batting first has won the match

batting_first = ipl[ipl['win_by_runs']!=0]
batting_first
```

```python
plt.figure(figsize = (7,7))
plt.hist(batting_first['win_by_runs'])
plt.xlabel("runs")
plt.ylabel("no of times")
plt.title("Distribution of Runs")
plt.show()
```

```python
# x axis has number of instances and y has the team won by number of runs
```

```python
batting_first['winner'].value_counts()
```

```python
f = ['MumbaiIndians',                 
'ChennaiSuperKings',            
'KingsXIPunjab']

o = [52,37,38]

# with most wins after batting first
plt.figure(figsize = (7,7))
plt.bar(f,o,color=['blue','yellow','red'])
```

```python
plt.figure(figsize = (5,5))
plt.pie(list(batting_first['winner'].value_counts()) ,labels= list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
# Extracting those records where team has won after batting second




batting_second = ipl[ipl['win_by_wickets']!=0]
batting_second
```

```python
batting_second.shape
```

```python
# make the histogram for frequency of wins w.r.t to number of wickts

plt.figure(figsize = (5,5))
plt.hist(batting_second['win_by_wickets'],bins = 30)
plt.xlabel("wickets")
plt.ylabel('no of matches')
plt.show()
```
